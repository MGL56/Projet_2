#### Idée : Scanner n'importe quel produit et obtenir un indicateur sur la qualité nutritionelle du produit

- Un modèle est entrainé sur les produits qui possèdent un Nutri-Score (X = macro-nutriments, y= Nutri-Score)
- Le modèle entrainé sert ensuite à prédire le Nutri-Score des produits qui n'en possèdent pas
- Les Nutri-Score prédits remplacent les valeurs manquantes dans la base de donnée
- L'utilisateur scanne un produit quelconque et obtient un score renseignant sa qualité nutritionelle

#### Informations sur le Nutri-Score :

Le Nutri-Score, aussi appelé système 5 couleurs, est un système d'étiquetage nutritionnel basé sur un logo avec cinq valeurs allant de A à E et du vert au rouge, établi en fonction de la valeur nutritionnelle d'un produit alimentaire. Il est mis en place par le gouvernement Français en 2016 dans le cadre de la loi de modernisation du système de santé.

Le score est calculé par un système de points, le score le plus faible étant le meilleur :

Eléments défavorables au score :
- apport calorique
- teneur en sucre
- teneur en graisses saturés
- teneur en sel

Eléments favorables au score :
- teneur en fruits, légumes, légumineuses et oléagineux
- teneur en fibres
- teneur en protéines

Des points sont attribués en fonction des éléments favorables et défavorables et un grade est attribué à chaque produit en fonction de son score



Pour plus d'informations : https://quoidansmonassiette.fr/comment-est-calcule-le-nutri-score-logo-nutritionnel/

### 2.2 Corrélations entre macronutriments et Nutri-Score

```python
# Set de données pour la création de modèles de machine learning
y_set = df["nutrition_grade_fr"].dropna()
x_set = df.dropna(subset=["nutrition_grade_fr"])[macro_nutriments_labels]


# On encode les labels [a,b,c,d,e] -> [0,1,2,3,4]
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y_set)

# On normalize les variables numériques
from sklearn.preprocessing import StandardScaler
encoder_x = StandardScaler()
x_set_encoded = encoder_x.fit_transform(x_set)


final_data = pd.DataFrame(x_set, columns = macro_nutriments_labels)
final_data["nutrition_grade_fr"] = y_encoded

# Matrice de corrélation des variables sélectionnés 
corr_matrix = final_data.corr()
# Affiche les résultats de corrélation sous forme de heatmap
fig = plt.figure(1, figsize=(10, 10))
plt.title("Corrélation des macro nutriments")
sns.heatmap(corr_matrix)
plt.show()
```

Les graisses et graisses saturées sont corrélées entre elles et avec le Nutri-Score. Le Nutri-Score est également corrélé avec le sucre et les calories

## 3 - Machine Learning : Sélection et entrainnement de modèles

Séparation du jeu de données en données d'entrainement et de test

```python

# On divise le jeux de données en 2 avec train_size = 80%
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_set_encoded, y_encoded, stratify=y_encoded ,train_size=0.8, random_state=42)
```

La séparation est faite de telle sorte que les proportions de chaque classe sont égales pour le jeu d'entrainement et de test

Le score de précision est le rapport de vrais positifs sur le nombre de faux positifs et vrais positifs :

Exemple classe A : 
 - Vrai positif : donnée prédite A et réellement A
 - Faux positif : donnée prédite A mais n'est pas A
 
Le score de précision est établit pour chaque classe et ensuite une moyenne est faite sur l'ensemble des classes pour obtenir un score entre 0 et 1

- Plusieurs modèles sont testés, chacun avec une cross validation à 10 passes, les scores sont ensuite comparés pour faire une séléction du modèle le plus performant
- Le meilleur modèle est ensuite hyperparamétrisé pour affiner la précision du modèle

### 3.1 Classifieur k-NN

```python
accuracy = {}
```

```python
# On entraine un modèle k-NN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

knn_clf = KNeighborsClassifier()

accuracy["knn"] = cross_val_score(knn_clf, x_train, y_train, cv=10)


print("Accuracy for a k-NN on validation dataset : {:.2f} %".format(100*accuracy["knn"].mean()))
```

### 3.2 Classifieur Random Forest

```python
# On entraine un modèle Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

forest_clf = RandomForestClassifier(n_estimators=20, random_state=42)

accuracy["random_forest"] = cross_val_score(forest_clf, x_train, y_train, cv=10)


print("Accuracy for a Random Forest Algorithm on validation dataset : {:.2f} %".format(100*accuracy["random_forest"].mean()))
```

### 3.3 Classifieur Boosting de gradient

```python
# On entraine un modèle Random Forest
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score

gb_clf = GradientBoostingClassifier(n_estimators=50)

accuracy["gradient_boosting"] = cross_val_score(gb_clf, x_train, y_train, cv=10)


print("Accuracy for a Boosting Gradient on validation dataset : {:.2f} %".format(100*accuracy["gradient_boosting"].mean()))
```

### 3.4 Séléction du meilleur modèle

Moyenne et ecart-type de chaque modèle sur une cross validation à 10 passes

```python
accuracy_mean = [accuracy.mean() for accuracy in np.array(list(accuracy.values()))]
accuracy_std = [accuracy.std() for accuracy in np.array(list(accuracy.values()))]
x = list(accuracy)     

fig = plt.figure(1, figsize=(10,5))
plt.subplot(211)
sns.barplot(x=x, y=accuracy_mean)
plt.ylabel("Moyenne")
plt.title("Accuracy on cross validation data")
plt.subplot(212)
sns.barplot(x=x, y=accuracy_std)
plt.ylabel("Ecart-Type")

plt.show()
```

```python
print("La performance du modèle k-neighbors est d'environ : {:.2f} % +- {:.2f} %".format(100*accuracy["knn"].mean(), 100*accuracy["knn"].std()))
print("La performance du modèle Random Forest est d'environ : {:.2f} % +- {:.2f} %".format(100*accuracy["random_forest"].mean(), 100*accuracy["random_forest"].std()))
print("La performance du modèle Gradient Boosting est d'environ : {:.2f} % +- {:.2f} %".format(100*accuracy["gradient_boosting"].mean(), 100*accuracy["gradient_boosting"].std()))
```

- Le meilleur modèle est celui qui obtient le score de précision le plus élevé en moyenne sur les 10 passes et qui présente la plus petite variance

- Le meilleur modèle est donc le Random Forest !

### 3.4 Réglage des hyperparamètres du classifier Random Forest

GridSearchCV permet de tester une grille d'hyperparamètres et de sélectionner le meilleur modèle

```python
# On améliore le meilleur modèle en réglant les meilleurs hyperparamètres
from sklearn.model_selection import GridSearchCV

# Grille de paramètres
param_grid = [{'max_features': [5, 6, 7, 8], 'max_depth': range(15, 25)}]

# Recherche par quadrillage des meilleurs paramètres
grid_search = GridSearchCV(forest_clf, param_grid, cv=5)
grid_search.fit(x_train, y_train)
grid_search.best_params_
```

```python
from sklearn.metrics import accuracy_score
# On récupère le meilleur modèle
best_clf = grid_search.best_estimator_

# Predictions sur les données test
y_pred = best_clf.predict(x_test)

# Précision du modèle final sur le jeu test
accuracy = accuracy_score(y_pred, y_test)
print("Accuracy for the best model on Test dataset : {:.2f} %".format(100*accuracy))

# On entraine le meilleur modèle sur tout le jeu de données disponibles
best_clf.fit(x_set_encoded, y_encoded)
```

#### Features importances

```python
feature_importances = grid_search.best_estimator_.feature_importances_
attributes = macro_nutriments_labels
# Les features sont triés par importance pour l'affichage graphique
x, y = [], []
for x_0, y_0 in sorted(zip(feature_importances, attributes), reverse=True) :
    x.append(x_0)
    y.append(y_0)
# Affichage Bar Plot
sns.barplot(x=x, y=y)
plt.ylabel("Macro Nutriments")
plt.xlabel("Feature importances")
plt.xticks(rotation=45)
plt.show()
```

- La recherhe de paramètres par quadrillage a séléctionné un modèle qui exclu les 2 dernières features (fat et carbohydrates), ce sont celles qui n'interviennnet pas dans le calcul du Nutri-Score, i.e les graisses et les glucides

### 3.4 Analyse des erreurs

#### 3.4.1 Matrice de confusion

- Les lignes représentent les véritables classes, tandis que les colonnes représentent les classes prédites
- Le score de précision : True positif / (True Positif + False Positif)

```python
from sklearn.metrics import confusion_matrix
conf_mx = confusion_matrix(y_test, y_pred)

row_sums = conf_mx.sum(axis=0, keepdims=True)
norm_conf_mx = conf_mx / row_sums *100

# On remplace les valeurs de la diagonale par des 0
np.fill_diagonal(norm_conf_mx, 0)

sns.heatmap(norm_conf_mx, cmap=plt.cm.gray, annot=True)
plt.title("Matrice de confusion (%)")
plt.xlabel("classes prédites")
plt.ylabel("classes réelles")
plt.xticks(np.arange(5), ('A', 'B', 'C', 'D', 'E'))
plt.yticks(np.arange(5), ('A', 'B', 'C', 'D', 'E'))

plt.show()
```

```python
print("Le score de précision est : \n \
      - la classe A : {:.2f} %\n \
      - la classe B : {:.2f} %\n \
      - la classe C : {:.2f} %\n \
      - la classe D : {:.2f} %\n \
      - la classe E : {:.2f} %\n".format(conf_mx[0,0]/conf_mx.sum(axis=0)[0]*100,
                                      conf_mx[1,1]/conf_mx.sum(axis=0)[1]*100,
                                      conf_mx[2,2]/conf_mx.sum(axis=0)[2]*100,
                                      conf_mx[3,3]/conf_mx.sum(axis=0)[3]*100,
                                      conf_mx[4,4]/conf_mx.sum(axis=0)[4]*100,
                                   ))

print("Parmi les produits prédits en classe A, seulement {:.2f} % sont réellement de la classe A \n \
{:.2f} % sont de la classe B, {:.2f} % sont de la classe C, {:.2f} % de la classe D et {:.2f} % de la classe E \n".format(conf_mx[0,0]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[1,0]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[2,0]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[3,0]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[4,0]/conf_mx.sum(axis=0)[0]*100))
     

print("Parmi les produits réellement de la classe A, seulement {:.2f} % sont prédits de la classe A \n \
{:.2f} % de la classe B, {:.2f} % de la classe C, {:.2f} % de la classe D et {:.2f} % de la classe E".format(conf_mx[0,0]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[0,1]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[0,2]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[0,3]/conf_mx.sum(axis=0)[0]*100,
                                                                                                                      conf_mx[0,4]/conf_mx.sum(axis=0)[0]*100))
```

Les taux d'erreurs les plus importants se situent entre classes voisines. Ex : B prédit en C.
Peu de produits sont prédits A alors qu'ils sont D ou E

#### 3.4.2 Ecart entre modèle et réalité

Il existe une certaine relation d'ordre (A>B>C>D>E) entre les scores ( Ex : A plus proche de B que de C)

```python
ecart = [list(abs(y_test - y_pred)).count(k)/len(y_test)*100 for k in range(5)]
plt.bar(height = ecart, x =range(5))
plt.xlabel("Ecart de prédiction")
plt.ylabel("% de produits")
plt.title("Proportions de l'écart du Nutri-Score")
plt.show()
```

- Lorsque la prédiction est fausse, elle est majoritairement de la classe "adjacente"

### 3.6 Sauvegarde des résultats

#### Remplacement des valeurs manquantes du Nutri-Score

Les résultats obtenus par notre meilleur modèle remplacent les valeurs manquantes du Nutri-Score

```python
# On remplace les valeurs manquantes de nutrition score par la prédiction du modèle
x_na = df[df.nutrition_grade_fr.isna()][macro_nutriments_labels]

y_na = best_clf.predict(x_na)
y_na = encoder.inverse_transform(y_na)
```

Affichage de la proportion de chaque classe Nutri-Score

```python
fig = plt.figure(1, figsize=(8,8))

# Affichage les valeurs réelles Nutri-Score
plt.subplot(211)
plt.title('Nutri-Score des produits')
df['nutrition_grade_fr'].dropna().value_counts(normalize=True).sort_index().plot(kind='pie', colors=colors, labels=labels)


# Remplacement des valeurs manquantes par les valeurs prédites du modèle
df["nutrition_grade_fr"][df["nutrition_grade_fr"].isna()] = y_na

# Affichage des Nutri-Score + résultats modèle
plt.subplot(212)
plt.title('Nutri-Score et résultats du modèle')
df["nutrition_grade_fr"].value_counts(normalize=True).sort_index().plot(kind='pie', colors=colors, labels=labels)

plt.show()
```

- Il y a une augmentation significative des produits de mauvaise qualité (E) au détriment d'une diminution importante des produits de meilleurs qualités (A) et (B) : cela peut s'expliquer par le fait que les entreprises ne souhaitent en général pas afficher le Nutri-Score de leurs produits mal notés. En effet le Nutri-Score ne peut pas être imposé aux industriels en raison du règlement européen de 2011 concernant l’information des consommateurs sur les denrées alimentaires. Les produits dont le Nutri-Score n'est pas renseigné est probablement un produit de mauvaise qualité nutritionnelle
